```python
import pandas as pd
from sqlalchemy import create_engine
import os
import csv
import numpy as np
import psycopg2
```

### Extract CSVs into DataFrames

```python
# File tw Load (Remember to Change These)
file_to_load_2019h = "Resources/2019-Hydro.csv"
file_to_load_2019w = "Resources/2019-Wind.csv"
file_to_load_2019s = "Resources/2019-Solar.csv"
file_to_load_2018h = "Resources/2018-Hydro.csv"
file_to_load_2018w = "Resources/2018-Wind.csv"
file_to_load_2018s = "Resources/2018-Solar.csv"
file_to_load_2017h = "Resources/2017-Hydro.csv"
file_to_load_2017w = "Resources/2017-Wind.csv"
file_to_load_2017s = "Resources/2017-Solar.csv"
file_to_load_solar_rating = "Resources/solar_rating.csv"
```

```python
# Read csv
hydro_2019 = pd.read_csv(file_to_load_2019h)
wind_2019 = pd.read_csv(file_to_load_2019w)
solar_2019 = pd.read_csv(file_to_load_2019s)
hydro_2018 = pd.read_csv(file_to_load_2018h)
wind_2018 = pd.read_csv(file_to_load_2018w)
solar_2018 = pd.read_csv(file_to_load_2018s)
hydro_2017 = pd.read_csv(file_to_load_2017h)
wind_2017 = pd.read_csv(file_to_load_2017w)
solar_2017 = pd.read_csv(file_to_load_2017s)
solar_rating = pd.read_csv(file_to_load_solar_rating)
```

```python
# create df with only required info
#hydro 17 - 19
hydro_2019_cl = hydro_2019[["Small Unit Installation Postcode", "Installations Quantity Total", "SGU Rated Output In kW Total"]]
hydro_2019_df = hydro_2019_cl

hydro_2018_cl = hydro_2018[["Small Unit Installation Postcode", "Installations Quantity Total", "SGU Rated Output In kW Total"]]
hydro_2018_df = hydro_2018_cl

hydro_2017_cl = hydro_2017[["Small Unit Installation Postcode", "Installations Quantity Total", "SGU Rated Output In kW Total"]]
hydro_2017_df = hydro_2017_cl
```

```python
# df with only required info
# Wind 17 - 19
wind_2019_cl = wind_2019[["Small Unit Installation Postcode", "Installations Quantity Total", "SGU Rated Output In kW Total"]]
wind_2019_df = wind_2019_cl

wind_2018_cl = wind_2018[["Small Unit Installation Postcode", "Installations Quantity Total", "SGU Rated Output In kW Total"]]
wind_2018_df = wind_2018_cl

wind_2017_cl = wind_2017[["Small Unit Installation Postcode", "Installations Quantity Total", "SGU Rated Output In kW Total"]]
wind_2017_df = wind_2017_cl
```

```python
# df with only required info
# Solar 17 - 19
solar_2019_cl = solar_2019[["Small Unit Installation Postcode", "Installations Quantity Total", "SGU Rated Output In kW Total"]]
solar_2019_df = solar_2019_cl

solar_2018_cl = solar_2018[["Small Unit Installation Postcode", "Installations Quantity Total", "SGU Rated Output In kW Total"]]
solar_2018_df = solar_2018_cl

solar_2017_cl = solar_2017[["Small Unit Installation Postcode", "Installations Quantity Total", "SGU Rated Output In kW Total"]]
solar_2017_df = solar_2017_cl
```

```python
# raname hydro
hydro_2019_df = hydro_2019_df.rename(columns={           "Small Unit Installation Postcode": "Postcode",
                                                         "Installations Quantity Total": "Installations Hydro 2019",
                                                         "SGU Rated Output In kW Total": "Output In kW Hydro 2019"})

hydro_2018_df = hydro_2018_df.rename(columns={           "Small Unit Installation Postcode": "Postcode",
                                                         "Installations Quantity Total": "Installations Hydro 2018",
                                                         "SGU Rated Output In kW Total": "Output In kW Hydro 2018"})

hydro_2017_df = hydro_2017_df.rename(columns={           "Small Unit Installation Postcode": "Postcode",
                                                         "Installations Quantity Total": "Installations Hydro 2017",
                                                         "SGU Rated Output In kW Total": "Output In kW Hydro 2017"})
```

```python
# rename wind
wind_2019_df = wind_2019_df.rename(columns={           "Small Unit Installation Postcode": "Postcode",
                                                         "Installations Quantity Total": "Installations Wind 2019",
                                                         "SGU Rated Output In kW Total": "Output In kW Wind 2019"})

wind_2018_df = wind_2018_df.rename(columns={           "Small Unit Installation Postcode": "Postcode",
                                                         "Installations Quantity Total": "Installations Wind 2018",
                                                         "SGU Rated Output In kW Total": "Output In kW Wind 2018"})

wind_2017_df = wind_2017_df.rename(columns={           "Small Unit Installation Postcode": "Postcode",
                                                         "Installations Quantity Total": "Installations Wind 2017",
                                                         "SGU Rated Output In kW Total": "Output In kW Wind 2017"})
```

```python
# rename Solar
solar_2019_df = solar_2019_df.rename(columns={           "Small Unit Installation Postcode": "Postcode",
                                                         "Installations Quantity Total": "Installations Solar 2019",
                                                         "SGU Rated Output In kW Total": "Output In kW Solar 2019"})

solar_2018_df = solar_2018_df.rename(columns={           "Small Unit Installation Postcode": "Postcode",
                                                         "Installations Quantity Total": "Installations Solar 2018",
                                                         "SGU Rated Output In kW Total": "Output In kW Solar 2018"})

solar_2017_df = solar_2017_df.rename(columns={           "Small Unit Installation Postcode": "Postcode",
                                                         "Installations Quantity Total": "Installations Solar 2017",
                                                         "SGU Rated Output In kW Total": "Output In kW Solar 2017"})
```

### Merging DataFrame By Year

```python
#Hydro data 2017-2019
Hydro_1 = pd.merge(hydro_2017_df,hydro_2018_df, on = "Postcode", how = "left")
Hydro = pd.merge(Hydro_1,hydro_2019_df, on = "Postcode", how = "left")
```

```python
#  Wind data 2017-2019
Wind_1 = pd.merge(wind_2017_df,wind_2018_df, on = "Postcode", how = "left")
Wind = pd.merge(Wind_1,wind_2019_df, on = "Postcode", how = "left")
```

```python
# solar data 2017-2019
Solar_1 = pd.merge(solar_2017_df,solar_2018_df, on = "Postcode", how = "left")
Solar = pd.merge(Solar_1,solar_2019_df, on = "Postcode", how = "left")
```

```python
# Replacing NaN with 0
Hydro = Hydro.replace(np.NaN, 0)
Hydro
```

```python
#  Replacing NaN with 0
Wind = Wind.replace(np.NaN, 0)
Wind
```

```python
#  Replacing NaN with 0
Solar = Solar.replace(np.NaN, 0)
Solar
```
